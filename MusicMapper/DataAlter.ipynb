```python
import os
import pandas as pd
import numpy as np
```

```python
folder = os.path.join(os.getcwd(), "data")
file = os.path.join(folder, "metal_bands_2017.csv")
data = pd.read_csv(file, encoding="unicode_escape")
data = data.drop("Unnamed: 0", axis=1)
data = data.drop("split",axis = 1)
```

```python
for row in data.itertuples():

  list_genres = row[5].split(",")
  band_name = row[1].split(",")
  country_name = row[4].split(",")
  genres = "_".join(list_genres)
  band = "_".join(band_name)
  country = "_".join(country_name)
  data.loc[row[0],"style"] = genres
  data.loc[row[0],"band_name"] = band
  data.loc[row[0],"origin"] = country
```

```python
data = data.set_index("band_name")
data.to_csv("data/metal_bands_2017_updated.csv")
```
